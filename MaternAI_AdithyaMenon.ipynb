```python
# Importing required libraries
import pandas as pd
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Load the dataset
df = pd.read_csv('X:/Predicting Pregancy Risks/cleaned_data1.csv')

# Splitting the dataset into features (X) and target (y)
X = df.drop(columns=['Outcome'])  # Assuming 'Outcome' is the target column
y = df['Outcome']

# Splitting data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Standardizing the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Encode the target labels
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Initialize the Classifiers
rf_model = RandomForestClassifier(random_state=42)
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
knn_model = KNeighborsClassifier(n_neighbors=5)
logreg_model = LogisticRegression(random_state=42, max_iter=10000)
gb_model = GradientBoostingClassifier(random_state=42)

# Train the models
rf_model.fit(X_train_scaled, y_train_encoded)
xgb_model.fit(X_train_scaled, y_train_encoded)
knn_model.fit(X_train_scaled, y_train_encoded)
logreg_model.fit(X_train_scaled, y_train_encoded)
gb_model.fit(X_train_scaled, y_train_encoded)

# Make predictions
y_pred_rf = rf_model.predict(X_test_scaled)
y_pred_xgb = xgb_model.predict(X_test_scaled)
y_pred_knn = knn_model.predict(X_test_scaled)
y_pred_logreg = logreg_model.predict(X_test_scaled)
y_pred_gb = gb_model.predict(X_test_scaled)

# Decode predictions back to original labels
y_pred_rf_decoded = label_encoder.inverse_transform(y_pred_rf)
y_pred_xgb_decoded = label_encoder.inverse_transform(y_pred_xgb)
y_pred_knn_decoded = label_encoder.inverse_transform(y_pred_knn)
y_pred_logreg_decoded = label_encoder.inverse_transform(y_pred_logreg)
y_pred_gb_decoded = label_encoder.inverse_transform(y_pred_gb)

# Calculate accuracy scores
accuracy_rf = accuracy_score(y_test_encoded, y_pred_rf)
accuracy_xgb = accuracy_score(y_test_encoded, y_pred_xgb)
accuracy_knn = accuracy_score(y_test_encoded, y_pred_knn)
accuracy_logreg = accuracy_score(y_test_encoded, y_pred_logreg)
accuracy_gb = accuracy_score(y_test_encoded, y_pred_gb)

# Print results
print("Random Forest Accuracy:", accuracy_rf)
print("XGBoost Accuracy:", accuracy_xgb)
print("KNN Accuracy:", accuracy_knn)
print("Logistic Regression Accuracy:", accuracy_logreg)
print("Gradient Boosting Accuracy:", accuracy_gb)

# Print decoded predictions for Random Forest as an example
print("Decoded Predictions (Random Forest):", y_pred_rf_decoded)
```

```python
# Define a function to map the outcome to risk levels
def risk_level(outcome):
    if outcome == 0:
        return "Low Risk"
    elif outcome == 1:
        return "Mid Risk"
    elif outcome == 2:
        return "High Risk"
    else:
        return "Unknown Risk"

# Function to take user input and predict risk level
def predict_risk():
    print("\nPlease enter the following details:")
    age = float(input("Age: "))
    body_temp = float(input("Body Temperature (F): "))
    heart_rate = float(input("Heart rate (bpm): "))
    systolic_bp = float(input("Systolic Blood Pressure (mm Hg): "))
    diastolic_bp = float(input("Diastolic Blood Pressure (mm Hg): "))
    bmi = float(input("BMI (kg/m^2): "))
    blood_glucose_hba1c = float(input("Blood Glucose (HbA1c): "))
    blood_glucose_fasting = float(input("Blood Glucose (Fasting hour-mg/dl): "))

    # Combine inputs into a single feature array (2D array for prediction)
    user_input = [[
        age,
        body_temp,
        heart_rate,
        systolic_bp,
        diastolic_bp,
        bmi,
        blood_glucose_hba1c,
        blood_glucose_fasting
    ]]
    user_input_scaled = scaler.transform(user_input)

    # Predict the outcome using the XGBoost model
    predicted_outcome = xgb_model.predict(user_input_scaled)[0]

    # Print the risk level based on the predicted outcome
    print(f"Predicted Risk Level: {risk_level(predicted_outcome)}")

# Main loop for multiple inputs
while True:
    print("\n--- Risk Prediction Menu ---")
    print("1. Enter new data for prediction")
    print("2. Exit")
    choice = input("Choose an option (1 or 2): ")

    if choice == '1':
        predict_risk()
    elif choice == '2':
        print("Exiting the program. Thank you!")
        break
    else:
        print("Invalid choice. Please select 1 or 2.")
```
